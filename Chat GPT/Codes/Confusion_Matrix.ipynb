```python
import ast
import json
import pandas as pd
```

```python
# Read File
file_name = "MaSaC_val_erc.json"
with open(file_name, 'r') as file:
        dataset = json.load(file)
# Number of episodes in dataset

N = len(dataset)
print(N)
# Reading the data

episodes_list = []
speakers_list = []
utterances_list = []
emotions_list = []

for i in range(len(dataset)):
        episodes_list.append(dataset[i]['episode'])
        speakers_list.append(dataset[i]['speakers'])
        utterances_list.append(dataset[i]['utterances'])
        emotions_list.append(dataset[i]['emotions'])  
```

```python
# Evaluation
# Open Response File
f = open("Task 1.txt", "r")

# Make Matrix
labels = ["neutral", "anger", "surprise", "fear", "joy", "contempt", "sadness", "disgust","null"]
dict = {}
for label1 in labels:
    for label2 in labels:
        dict[label1,label2] = 0

# Calculate
for i in range(len(dataset)):
    predictions = ast.literal_eval(f.readline())
    for j in range(len(utterances_list[i])):
        true_emotion = emotions_list[i][j]
        try:
            pre_emotion = predictions[j]
        except:
            print("Emotion not predicted for ",(i,j))
            pre_emotion = "null"
        dict[pre_emotion, true_emotion] += 1
        
# Displaying Results
result = {}
for label1 in labels:
    result[label1] = [0] * len(labels)

for i, label1 in enumerate(labels):
    for j, label2 in enumerate(labels):
        result[label1][j] = dict[label1, label2]

df_count = pd.DataFrame(result, pd.Index(labels))
df_count.to_csv("Confusion_Matrix.csv")
```
